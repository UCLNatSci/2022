```python
import numpy as np
import matplotlib.pyplot as plt

def height(x, y):
    val = np.exp(-y**2)      
    if x < 1.5 and x > -1.5:
        result = val
    else:
        result = 0    
    return result

# check the function works
h = height(0, 0)
print(h) # should print 1

h = height(-1.5, 0)
print(h) # should print 0
```

```python
N = 5
z = np.zeros((N, N))

x_array = np.linspace(-2, 2, N)
print("x_array:", x_array)

for i in range(N):
    x = x_array[i]
    z[i,0] = height(x, 0)
    
print("sphere:\n", z)
```

```python
N = 30

z = np.zeros((N, N))

x_array = np.linspace(-2, 2, N)
print("x_array:", x_array)
y_array = np.linspace(-2, 2, N)
print("y_array:", y_array)

for j in range(N):
    for i in range(N):
        x = x_array[i]
        y = y_array[j]
        z[i,j] = height(x, y)
    
print("z:\n", z)

plt.imshow(z)
```

```python
cross_section_0 = z[:,12]

plt.plot(x_array, cross_section_0)
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')


# Plot the surface
x1, y1 = np.meshgrid(x_array, y_array)
ax.plot_surface(x1, y1, z)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img

def funky_func(x,y):
    return 3*(1-x)**2*np.exp(-x**2-(y+1)**2)-10*(x/5-x**3-y**5)*np.exp(-x**2-y**2)-(1/3)*np.exp(-(x+1)**2-y**2)

N = 94


z = np.zeros((N, N))

x_array = np.linspace(-3, 3, N)
print("x_array:", x_array)
                  
y_array = np.linspace(-3, 3, N)
print("y_array:", y_array)

for i in range(N):
    for j in range(N):
        x = x_array[i]
        y = y_array[j]
                  
        z[i,j] = funky_func(x, y) * 2.063 + 1010 
    


plt.imshow(z, )
plt.axis("off")

fig = plt.figure(figsize=(6,6))
x1, y1 = np.meshgrid(x_array, y_array)
ax = fig.add_subplot(projection='3d')


# Plot the surface
ax.plot_surface(x1, y1, z)

# Set an equal aspect ratio
#ax.set_aspect('equal')

img.imsave("pressure_2d.png", z, cmap="gray")
np.savetxt("pressure_2d.txt", z)
```
