```python
import numpy as np
import matplotlib.pyplot as plt

x = np.loadtxt("falling_cat.txt")

n, m = x.shape

k = 4

n_out = n // 2
m_out = m // 2

y = np.zeros((n_out, m_out))

for i in range(n_out):
    for j in range(m_out):
        z = np.mean(x[i*2:i*2+2,j*2:j*2+2])
        y[i, j] = z


plt.imshow(y, cmap="gray")
```
